# Validation set EDA

```python
category_valsub1=dict()
category_valmajor1=dict()
category_valsub2=[]
category_valmajor2=[]
category_valmasub=dict()
category_valmasub2=[]
for dirpath, dirnames,filenames in os.walk(os.curdir):

    for filename in filenames :
#         print(os.path.join(dirpath,filename))
#         print(dirpath.split('/'))
        cate=dirpath.split('/')
        category_valsub1[cate[2]]=category_valsub1.get(cate[2],0)+1
        category_valmajor1[cate[1]]=category_valmajor1.get(cate[1],0)+1
        category_valsub2.append(cate[2])
        category_valmajor2.append(cate[1])
        category_valmasub[(cate[1],cate[2])]=category_valmasub.get((cate[1],cate[2]),0)+1
        category_valmasub2.append((cate[1],cate[2]))
#         train_category.append(os.path.join(dirpath,filename))
```

```python
category_totalmasub1=category_valmasub2+category_trainmasub2
```

```python
category_totalmasub2=dict()
for key in category_totalmasub1:
    category_totalmasub2[key]=category_totalmasub2.get(key,0)+1
```

```python
subcate=[]
count=[]
for key, value in category_totalmasub2.items():

    if key[0]=='의류' :
#         print(key[1], value)
        subcate.append(key[1])
        count.append(value)
    else :
        continue
data={"의류" :subcate,
     "이미지 수" : count}

df = pd.DataFrame.from_dict(data)
df
```

```python
fig, ax1 = plt.subplots(figsize=(10, 10))
plt.rcParams['font.family'] = font
print(plt.rcParams['font.family'])
sns.barplot(x='의류', y='이미지 수',data=df, ax=ax1)
sns.despine(fig)
```
